```python
# create rich menu - json
import json

import requests


with open('env.json', encoding='utf-8') as f:
    env = json.load(f)

url = 'https://api.line.me/v2/bot/richmenu'

headers = {'Authorization': f'Bearer {env['CHANNEL_ACCESS_TOKEN']}'}

with open('navall.json', encoding='utf-8') as f:
    _json = json.load(f)

r = requests.post(url, headers=headers, json=_json)

# 'richmenu-5e5fd3ecae4f0ee760b4f885e065bd3d'  # navall.json
# 'richmenu-663c6b997ba713a8c2c33d7c0f672fd4'  # fans.json
# 'richmenu-113ebbf2fe7b01ad064b393d6b8791fb'  # rich1.json
# 'richmenu-15981903ae8a6ccc590b8b07e7881293'  # rich2.json
```

```python
r.json()
```

```python
# upload pic of rich menu

richmenuid = 'richmenu-5e5fd3ecae4f0ee760b4f885e065bd3d'
url = f'https://api-data.line.me/v2/bot/richmenu/{richmenuid}/content'
headers = {'Authorization': f'Bearer {env['CHANNEL_ACCESS_TOKEN']}',
           'Content-Type': 'image/jpeg'  # image/png
          }
with open('navall.jpg', 'rb') as f:
    data = f.read()

requests.post(url, headers=headers, data=data)
```

```python
# get rich menu list

url = 'https://api.line.me/v2/bot/richmenu/list'
headers = {'Authorization': f'Bearer {env['CHANNEL_ACCESS_TOKEN']}'}

r = requests.get(url, headers=headers)
```
